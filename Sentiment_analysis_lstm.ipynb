# integer encoding

```python
# from keras.datasets import imdb
from keras.preprocessing.text import Tokenizer
from keras.utils import pad_sequences
from keras import Sequential
from keras.layers import Dense,Embedding,Flatten, LSTM
```

```python
tokenizer = Tokenizer(oov_token='<nothing>')
```

```python
## taking only ist 10K documents

final_data = data['tweet']
```

```python
tokenizer.fit_on_texts(final_data) # DOING TOKEINZATION
```

```python
# Get the unique words (vocabulary)
len(tokenizer.word_index)
```

```python
sequences = tokenizer.texts_to_sequences(final_data)
sequences
```

# padding

```python
data_padded = pad_sequences(sequences,padding='post',maxlen=20)
data_padded[0:3]
```

```python
type(data_padded), data_padded.ndim, data_padded.shape
```

# DOing keras embedding just for demonstration but we will only pass the padded data in model as we are adding tbe keras embedding layer in the model , so no need to do seperate as it will create dimensionality issue

```python
model = Sequential()
model.add(Embedding(len(tokenizer.word_index),output_dim=2,input_length=20))  ## 23 = unique words in vocabulary len(tokenizer.word_index) =23 , out_dim =2 means higher dimension output the model generates after learning ,
                                                              ## input_length=5 means length of padding or sentences passing

model.summary()
```

```python
model.compile('adam','accuracy')
```

```python
pred = model.predict(data_padded)  ## 2D vector ggeenerated afyer leaning, this is not constant bcpoz its weights  will get updated internally with RNN, LSTM or GRU during back propogation in training
print(pred)
```

# Sentiment analysis

```python
from sklearn.model_selection import train_test_split

train_data, test_data, train_labels, test_labels = train_test_split(data_padded,labels, test_size=0.2, random_state=42)
```

```python
len(train_data), len(test_data), len(train_labels), len(test_labels)
```

```python
type(train_data), type(train_labels)
```

```python
train_data.shape
```

```python
train_data
```

```python
model = Sequential()
model.add(Embedding(input_dim=len(tokenizer.word_index) ,output_dim= 2,input_length=20))  ## this is not constant bcpoz its weights  will get updated internally with RNN, LSTM or GRU during back propogation in training
model.add(LSTM(32,return_sequences=False))
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
history = model.fit(train_data, train_labels,epochs=5,validation_data=(test_data,test_labels))
```

```python
pred_array = model.predict(test_data)
pred_array
```

```python
transformed_values = list(map(lambda val: 1 if val >= 0.5 else 0, pred_array))

# Count the occurrences of 0s and 1s
from collections import Counter

counted_values = Counter(transformed_values)

# Print the counts
print("negative sentiment for the tweet:", counted_values[0])
print("positive sentiment for the tweet:", counted_values[1])
```
