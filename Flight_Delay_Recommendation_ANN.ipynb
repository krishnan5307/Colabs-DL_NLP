```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Input


# Split the input features (X) and target variables (y)
X = data.drop(['CANCELLED', 'ACTUAL_ELAPSED_TIME','CARRIER_DELAY', 'WEATHER_DELAY', 'SECURITY_DELAY', 'LATE_AIRCRAFT_DELAY'], axis=1)
y_flight_duration = data['ACTUAL_ELAPSED_TIME']
y_cancelled = data['CANCELLED']
y_delay_reasons = data[['CARRIER_DELAY', 'WEATHER_DELAY', 'SECURITY_DELAY', 'LATE_AIRCRAFT_DELAY']]

# Split the data into training and testing sets
X_train, X_test, y_train_flight_duration,y_test_flight_duration, y_train_cancelled, y_test_cancelled, y_train_delay, y_test_delay = train_test_split(
    X,y_flight_duration, y_cancelled, y_delay_reasons, test_size=0.2, random_state=42
)

# Scale numerical features if necessary
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)          ## transfroming the test data with learning parameters of train data itself , so fit is not used inoder to avoid data leankage of test data
```

```python
X_train.shape
```

```python
X_train_scaled
```

```python
X_test_scaled
```

```python
y_delay_reasons
```

```python

# Define the ANN architecture
input_dim = X_train_scaled.shape[1]  ## gettunf column size
output_dim_flight_duration = 1
output_dim_cancelled = 1
output_dim_delay = y_train_delay.shape[1]

input_layer = Input(shape=(input_dim,))
hidden_layer = Dense(64, activation='relu')(input_layer)
# Add additional hidden layers if needed
output_layer_flight_duration = Dense(output_dim_flight_duration, activation= 'linear',name = 'flight_duration')(hidden_layer)
output_layer_cancelled = Dense(output_dim_cancelled, activation='sigmoid', name='cancelled')(hidden_layer)
output_layer_delay = Dense(output_dim_delay, activation='relu', name='delay_reasons')(hidden_layer)

model = Model(inputs=input_layer, outputs=[output_layer_flight_duration, output_layer_cancelled, output_layer_delay])

# Compile the model
model.compile(loss={'flight_duration':'mean_squared_error', 'cancelled': 'binary_crossentropy', 'delay_reasons': 'mean_squared_error'},
              optimizer='adam',
              metrics={'flight_duration':'mse','cancelled': 'accuracy', 'delay_reasons': 'mse'})


```

```python
model.summary()
```

```python
## Early Stopping
import tensorflow as tf
early_stopping=tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    patience=5

)
```

```python
X_train_scaled.shape
```

```python
y_train_delay
```

```python
# Train the model
model.fit(X_train_scaled, {'flight_duration':y_train_flight_duration, 'cancelled': y_train_cancelled, 'delay_reasons': y_train_delay},  ## fitting each output layer with respective target data for training model
          epochs=30,
          batch_size=24,
          validation_data=(X_test_scaled, {'flight_duration':y_test_flight_duration,'cancelled': y_test_cancelled, 'delay_reasons': y_test_delay}))#, callbacks=early_stopping)
```

```python
# Evaluate the model
losses = model.evaluate(X_test_scaled, {'flight_duration': y_test_flight_duration, 'cancelled': y_test_cancelled, 'delay_reasons': y_test_delay})

losses
```
